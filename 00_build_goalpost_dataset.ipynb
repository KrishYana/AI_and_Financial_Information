## 0) Notebook contract (constants + folders)
Define the dataset build configuration, create required folders, and print the resulting contract.

```python
from pathlib import Path

# Date bounds
START_DATE = "2018-01-01"
END_DATE = "2024-12-31"

# Universe settings
SECTOR_OPTIONS = ["Application Software", "Software—Infrastructure"]
SECTOR = "Application Software"
MIN_TRANSCRIPTS_PER_TICKER = 24

# Project paths
PROJECT_ROOT = Path.cwd()
DATA_DIR = PROJECT_ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
CACHE_DIR = DATA_DIR / "cache"
PARQUET_DIR = DATA_DIR / "parquet"
LOGS_DIR = DATA_DIR / "logs"

for path in [RAW_DIR, CACHE_DIR, PARQUET_DIR, LOGS_DIR]:
    path.mkdir(parents=True, exist_ok=True)

print("Notebook contract")
print("-" * 40)
print(f"START_DATE: {START_DATE}")
print(f"END_DATE: {END_DATE}")
print(f"SECTOR_OPTIONS: {SECTOR_OPTIONS}")
print(f"SECTOR: {SECTOR}")
print(f"MIN_TRANSCRIPTS_PER_TICKER: {MIN_TRANSCRIPTS_PER_TICKER}")
print("Created/verified folders:")
for path in [RAW_DIR, CACHE_DIR, PARQUET_DIR, LOGS_DIR]:
    print(f"  - {path}")
```
